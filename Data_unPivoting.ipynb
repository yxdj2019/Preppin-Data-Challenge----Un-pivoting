### Stack data

```python
df_new=df.stack().reset_index()
```

```python
df_new.head()
```

### Split column

```python
df_new.level_1.str.split('___', expand=True)
```

```python
df_new[['Bike_type','Date','Value']]=df_new.level_1.str.split('___', expand=True)
```

```python
df_new.drop(columns='level_1', inplace=True)
```

```python
df_new.head()
```

```python
# pd.Timestamp(df_new.Date.values)
```

### Tranform 'Date' to datetime type

```python
df_new.Date=pd.to_datetime(df_new.Date, format='%b_%y')
```

```python
df_new.head()
```

### Delete column level 0

```python
df_final=df_new.pivot_table(index=['Region','Bike_type','Date'],columns='Value').reset_index(col_level=1).droplevel(0, axis=1)
```

### Output

```python
df_final.to_csv('Output.csv', index=False)
```
